```python
import numpy as np
x = np.arange(20)
y = np.random.randint(20, size=20)
print('x : ',x)
print('y : ',y)
```

```python
#LineGraph
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))

plt.title('Graph',color='yellow',size=20)
plt.xlabel('X_Label',color='yellow',size=20)
plt.ylabel('Y_Label',color='yellow',size=20)
plt.legend()
plt.grid(True)

plt.plot(x,y, linestyle='dotted',marker='x',color='red',linewidth=5,markersize=20,label='First')

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))

plt.title('Graph',color='yellow',size=20)
plt.xlabel('X_Label',color='yellow',size=20)
plt.ylabel('Y_Label',color='yellow',size=20)
#plt.legend(['First Graph', 'Second Graph'] , loc='upper left')
plt.legend()
plt.grid(True)

plt.plot(x1,y1, linestyle='dotted',marker='x',color='red',linewidth=5,markersize=10, label = 'XY1')
plt.plot(x2,y2, linestyle='dotted',marker='o',color='green',linewidth=5,markersize=10, label = 'XY2')

plt.show()
```

```python
#Scatter Graph
plt.figure(figsize=(20,10))

plt.title('Graph',color='yellow',size=20)
plt.xlabel('X_Label',color='yellow',size=20)
plt.ylabel('Y_Label',color='yellow',size=20)
plt.legend()
plt.grid(True)

plt.scatter(x,y, color='red',marker = 'x', s=50, label='First')

plt.show()
```

```python
#Bar Graph
x = np.arange(10)
y = np.random.randint(100, size=10)

plt.figure(figsize=(20,10))
plt.title('Graph',color='yellow',size=20)
plt.xlabel('X_Label',color='yellow',size=20)
plt.ylabel('Y_Label',color='yellow',size=20)
plt.legend('vipul')
plt.grid(True)


plt.bar(x, y, color='orange', label='Line')

plt.show()
```

```python
x1 = np.arange(10)
x2 = np.arange(10)

y1 = np.random.randint(100, size=10)
y2 = np.random.randint(100, size=10)

plt.bar(x1,y1)
plt.bar(x2,y2)

plt.title('Graph',color='yellow',size=20)
plt.xlabel('X_Label',color='yellow',size=20)
plt.ylabel('Y_Label',color='yellow',size=20)

plt.show()
```

```python
count = np.random.randint(100,size=(10,10))
columns = [2011+i for i in range(10)]

dt = pd.DataFrame(data=count, columns=columns)
months = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct']
dt.index = months
dt
```

```python
plt.figure(figsize=(20,20))
sns.set_context('paper', font_scale=3)
sns.heatmap(dt, vmin=0, vmax=100, annot=True, cmap='Blues', linewidth=2, linecolor='orange')
plt.title('HeatMap', color='blue')
plt.show()
```
